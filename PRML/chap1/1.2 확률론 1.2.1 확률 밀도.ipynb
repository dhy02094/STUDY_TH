# 확률론

불확실성에서 확률론을 이용하여 최적의 예측을 시행할 수 있다.  
  
**합의 법칙** $p(X) = \sum_Y p(X,Y)$   
**곱의 법칙** $p(X, Y) = p(Y|X)p(X)$  
  
## 베이즈정리(Bayes' theorem)

우선 곱의 법칙과 대칭성 $p(X,Y) = p(Y,X)$ 로부터  
$p(Y|X) = \frac{p(X|Y)p(Y)}{p(X)}$ 이 식이 베이즈정리   
베이지안 정리의 분모는 정규화 상수로 보면서 합의 법칙을 사용해 다음과 같이 표현 가능  
$p(X) = \sum_Y p(X|Y)p(Y)$  

여기서 **사전확률 : p(Y)** / **사후확률 : P(Y|X)** 이다.
X를 관측한 후의 확률이라 그렇다. 여기서 사후확률을 베이즈정리를 사용해서 구할 수 있는 것이다.  

### 독립일 때의 베이즈 정리
$p(X,Y) = p(X)p(Y)$ 처럼 독립인 경우를 고려해보자.  
곱의 법칙에 따라 $p(Y|X)=p(Y)$ 임을 알 수 있다.

# 1.2.1 확률 밀도
이산적인 변수에서의 확률에 대해 알아보도록 하자.  
 **확률밀도(probability density)** : 만일 $x$가 $(x, x + \delta x)$ 구간 안의 값을 가지고 그 변수의 확률이 $p(x)\delta x$로 주어지면 $p(x)$를 확률밀도라고 한다.   
 만일 $x$가 (a,b) 구간 사이의 값을 가질 확률은 
 $p(x \in (a,b)) = \int_a^b p(x) dx$ 
   - $p(x) \ge 0$
   - $\int_{-\infty}^{\infty} p(x) dx = 1 $

<img src = https://annajeong.github.io/assets/images/prml/figure1_12 width = '500' height = '500'/>

확률 분포 함수는 야코비안 인자로 인해서 비선형 변수 변환시 단순 함수와는 다른 방식으로 변한다.   
   - ex) $x = g(y)$의 변수 변환을 고려
       - f(x)는 $\tilde{f}(y) = f(g(y))$가 된다.
       - x의 확률 밀도 함수 $p_x(x)$와 새로운 변수 y의 확률 밀도 함수 $p_y(y)$를 살펴보면 둘이 다른 확률 밀도를 가진다는 것이 자명  
       - $(y, y + \delta y)$로 변환될것이다. 
       - $p_x(x)\delta x \simeq p_y(y)\delta y$
       - $p_y(y) = p_x(x)|\frac{dx}{dy}| = p_x(g(y))|g'(y)|$
   - 여기서 확률 밀도의 최댓값은 어떤 변수를 선택하는지에 따라 달라짐을 알 수 있다.
   
   
x가 $(-\infty, z)$ 범위에 속할 확률 => 누적 분포 함수(cumulative disributuion function)  
   - $P(z) = \int_{-\infty}^z p(x)dx$ $(P'(x) = p(x) )$
   - 만약 여러개의 연속적인 변수 $x_1,...,x_D$가 주어지고 이 변수들이 벡터 x로 표현될 경우 -> 결합확률밀도 $p(x) = p(x_1,...,x_D)$를 정의
   - 다음 조건을 만족해야 한다.
       - $p(x) \ge 0$
       - $\int p(x) dx = 1$
  
  
x가 이산 변수일 경우 p(x)를 종종 확률 질량 함수(probability mass function)이라고 부르기도 한다. 
- 각 x값에 대해서 몇몇 확률 질량들이 모여 있는 것으로 보기 때문

### 연속 변수의 확률 밀도와 이산 변수/연속 변수가 조합된 경우의 확률 밀도에도 합의 법칙, 곱의 법칙, 베이지안 정리 적용 가능
- ex) x와 y가 각각 실수와 변수일 경우, 합과 곱의 법칙은 다음의 형태를 띤다.
    - $p(x) = \int p(x,y)dy$
    - $p(x,y) = p(y|x)p(x)$
- 연속 변수의 합과 곱의 법칙에 대해서는 측도이론(measure theory)를 살펴야 하는데 책에 범위를 뛰어 넘어서 생략하도록 한다.
