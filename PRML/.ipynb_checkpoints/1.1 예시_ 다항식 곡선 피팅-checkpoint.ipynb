## 피팅

다음 sin(2$\pi$x) 함수를 활용해 데이터를 만들고, 약간의 랜덤한 노이즈 포함시켰다.


<img src = http://norman3.github.io/prml/images/Figure1.2.png width = '500' height = '500'/>  

  
이제 N개의 관찰값 $x$로 이루어진 훈련 집합 X $\equiv$ $(x_1, ..., x_N)^T$와 그에 해당하는 표적값 t $\equiv$ $(t_1, ..., t_N)^T$를 가정하자  
  
우리의 목표 : 훈련 집합들을 사용해서 새로운 입력값 x가 주어질 때 타깃변수 t를 예측 하는 것.  
  
앞으로 공부할 확률론 -> 불확실성을 정확하고 정량적으로 표현하는데 도움을 줄 것. + 1.5절 최적의 예측을 하는 확률로 표현. 
  
다시 본론으로 들어와 저 위 식을 피팅하는데 다음 다항식을 이용  
$y(x, w)$ = $w_0$ +$w_1$$x$ + $w_2$$x^2$ + ... + $w_M$$x^M$ = $\sum_{j=0}^M w_j x^j$  
  
M은 다항식의 차수  
다항함수 $y(x,w)$ 는 $x$에 대해서는 비선형이지만, $w$에 대해서는 선형이다.  
  
  
자세하게 설명할 수 있지만 일단은 넘어가겠다.  

## 오차함수

오차함수 : 훈련 집합의 표적값들의 값과 함수값 $y(x, w)$와의 오차를 측정
   - 다음 값을 최소화하는 방식으로 피팅해야한다. 
   - 가장 많이 쓰는 간단한 오차함수 사용
       - $E(w)$ = $1 \over 2$ $\sum_{n=1}^N$$(y(x_n,w) - t_n)^2$
       - (어떤 오차함수 인지, 왜 사용했는지는 나중에 설명)
   - E(w)를 최소화하는 w값을 선택함으로 곡선 피팅 문제 해결 가능 
       - 오차 함수를 계수에 대해 미분하면 w에 대한 선형식이 나올 것이다. -> 최소화하는 $w^*$을 찾을 수있다.

다음은 M = 0,1,3,9   
  
M=0,1 잘표현 하지 못했다.( 과소적합)
M=9 모든 부분은 지나가지만 심하게 진동하며 원함수를 표현 못했다.(과적합) 
  
M 값에 따라 일반화의 성능이 어떻게 변화는지 정량적으로 살펴보려면  
    => 평균 제곱근 오차(RMS error)  
    
    
$E_{RMS}$ = $\sqrt{2E(w^*)/N}$  

1. N으로 나눔으로써 데이터 사이즈가 다른 경우에도 비교 가능
2. 제곱근을 취함으로써 표적값 t와 같은 크기를 가지도록 했다.  
3.  <img src = http://norman3.github.io/prml/images/Figure1.5.png width = '400' height = '400'/>
4. M=3 이 가장 이상적인 값이고, M=9 일땐 훈련 오차가 0인것은 당연하고 시험에서 무진장 높아지므로 과적합이다.


## Q) 데이터의 크기 N이 달라진다면? 또 과적합을 해결하려면?
<img src = http://norman3.github.io/prml/images/Figure1.6a.png width = '300' height = '300'/>
<img src = http://norman3.github.io/prml/images/Figure1.6b.png width = '300' height = '300'/>  

   
   
1. 데이터의 집합의 수가 클수록 과적합 문제가 완화된다. (더 복잡한 모델을 활용해 피팅 가능)      
2. 과적합을 해결하는 방법에는 베이진안 방법을 채택하면 해결할 수 있다.(나중에 제대로 배워보자)      
3. 정규화(regularzation)     
    - 식 1.2의 오차 함수에 계수의 크기가 커지는 것을 막기 위한 페널티항을 추가하는 것.  
    - 가장 간단한 형태는 각각의 계수들을 제곱하여 합하는 것이다.  
    - L2 인  것 같다.  
    - $\tilde{E}(w) = {1\over2}\sum(y(x_n, w) - t_n)^2 + {\lambda \over 2}||w||^2$
    - (가끔 $w_0$은 정규화항에서 제외한다. 이걸 포함하면 타깃 변수의 원점을 무엇으로 선택하느냐에 따라 결과가 종속됨)
    - $\lambda$가 정규화항의 제곱합 오류항에 대한 상대적 중요도를 결정 한다.
    - 너무 큰 $\lambda$를 쓰지말고 이것이 모델의 복잡도를 조절해서 과적합 정도를 통제한다.

## 모델 복잡도를 잘 선택하는 단순한 방법
데이터집합을 훈련집합과검증집합으로 나누는 것이다. 
   - 훈련 집합 : 계수 w를 결정하는 데 활용
   - 검증 집합 : 모델 복잡도를 최적화하는데 활용(M이나 $\lambda$)
   
아주 간단하게 다항식 피팅을 하면서 생기는 문제에 대해 살펴보고 그것을 해결하는 방법에 대해 알아봤다. 뒤에서 이것에 대해 좀 더 자세히 익혀서 매끄럽게 쌩초보가 읽어도 해결할 수 있는 글을 작성해보자.
