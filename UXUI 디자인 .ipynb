# 1  UX/UI 디자인의 이해


- 개발자가 ux를 알아야 하는 이유?
    - 이걸 알면 담당자와 소통도 편하고 결과물은 개발 이 잘됐다는 증표같은 것이다.
    
- UX 란?
    - User Experience
    - 사용자가 사용하면서 겪는 경험

- UX 
    1. 사용자 경험을 통한 사용자들의 문제와 니즈 찾기
    2. 좋은 경험을 제공하기 위한 개선, 아이디어 도출
    3. 사용자에게 새로운 경험 제공
    
- UI 란?
    - User Interface
    - 사용자와 기기, 사용자와 웹 또는 어플이 만나는 접점으로 사용자가 보는 화면 디자인
    - ex) 갤럭시 one UI가 있고 나 같은 경우 아이폰 UI가 너무 이쁘고 심플해서 쭉 써오는 중
    
- UI 
    1. UX작업의 시각적 결과물
    2. 사용자의 편의성을 고려한 디자인
    3. 사용자의 행동을 유도하는 디자인
    
- UI가 빙산이라면 UX는 90%를 차지하고 있는 바다속 빙산이다.


- 음 UX같은 경우를 조금 더 쉽게 말하면 나같은 경우 어제 고용노동보험조회 좀 하려고 했는데 홈페이지에서 아무리 찾아도 찾기 힘들었다. 이런 경우 UI가 부족한거고 UX의 안 좋은 경험으로 남아야한다. 
- 반면 영화를 예매할 때는 부모님이 하시기에도 충분히 쉽게 예매할 수 있으므로 이런 경우 좋은 UX,UI 설정이 된 것 같고 UI,UX 가 중요한 이유까지도 설명될 듯 하다.

# 2. 어포던스 디자인 이해

- UX를 높이기 위한 UI 핵심 요소
    - 사용성, 편의성, 조작성
- 어포던스 디자인 = 행동유도성 디자인
- 지각 행동 유도성 : 사물이 쉽게 지각 될수 있어야하고 정체성이 해석 될 수 있어야 하고 무엇을 할 수 있고 해야 하는지 알아야 한다.
- UX/UI 원칙 
    1. 가시성
    2. 집중
    3. 반응
    4. 제약
    5. 맥락
    6. 일관성
    7. 움직임
    8. 예측가능
