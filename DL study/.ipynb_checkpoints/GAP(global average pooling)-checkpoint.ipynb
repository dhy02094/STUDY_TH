# Pooling  
  
- Pooling : CNN 과정을 거치다 보면 필터의 수가 많아져 차원이 높아집니다.
    - 따라서 필요한 파라미터의 개수는 많고 파라미터가 많다며 over fitting의 위험이 있어 줄여줄 방법이 필요합니다. 
    - 이것의 해결책이 pooling layer 입니다.
- 주로 Max Pooling을 주로 사용합니다. 

    - 1. 가장 큰 feature 값이 계산 시 가장 큰 영향을 주기 때문입니다.
    - 2. 특징을 가장 잘 나타내기 때문입니다. 
    - 3. stride = 2로 정하면 feautre의 크기를 반으로 줄 수 있다. 
  
- Average Pooling
    - 1. Max Pooling에서 최대값을 가져왔다면 Average Pooling에서는 평균값을 가져옵니다. 
    
# Global Average Pooling

- global 이라는 이름처럼 feature map 전부에 적용을 시킨다.  
- 앞서 설명한 pooling 보다 더 급격하게 feature의 수를 줄여줍니다. 
- 앞 pooling들과 목적이 다릅니다.
    - feature를 1차원 벡터로 만들기 위해서 사용합니다. 
    
- 같은 채널의 feature들을 모두 평균을 내서 채널의 갯수 만큼의 원소를 가지는 벡터로 만듭니다.   
    - ex) 7,7,512 의 output shape을 갖는다면 512개의 원소를 갖는 벡터가 된다.
- 왜? GAP를 사용할까?
    1. 보통의 경우 CNN+FC(Fully Connected) 에서 FC 에서 Layer에서 분류를 진행
    2. FC layer를 분류로 사용할 땐 파라미터의 수가 많이 증가하는 단점이 있다.
    3. 또한 feature 전체를 matrix 연산하기 때문에 위치에 대한 정보도 사라진다.
    4. input 크기도 특정 크기로 지정해야한다.
    5. 하지만 GAP는 어떤 크기의 feature라도 같은 채널의 값들을 하나의 평균값으로 대체하기 때문에 input은 중요하지 않아진다. 
    6. (H, W, C) -> (1, 1, C) 로 줄어드는 연산이므로 파라미터가 추가되지 않아 학습 측면에서 유리하다.
    7. GAP 연산 결과 1차원 벡터가 되기 때문에 최종 출력에 FC Layer 대신 사용가능.
    
    8. 위치정보를 지키기 때문에 CAM(Class Activation Mapping)을 사용해서 분류하는데 중요한 위치정보를 파악할 수 있어 사용한다.
