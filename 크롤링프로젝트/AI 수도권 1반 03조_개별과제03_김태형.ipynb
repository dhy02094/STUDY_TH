#### 1. 개별과제 : 키워드에 대한 검색량 데이터 수집

```python
# 추가로 필요한 import
import time
import random
import requests
```

```python
import time
import random
import requests

import signaturehelper

# 필요한 header 함수
def get_header(method, uri, api_key, secret_key, customer_id):
    timestamp = str(round(time.time() * 1000))
    signature = signaturehelper.Signature.generate(timestamp, method, uri, SECRET_KEY)
    return {'Content-Type': 'application/json; charset=UTF-8', 
            'X-Timestamp': timestamp, 'X-API-KEY': API_KEY, 
            'X-Customer': str(CUSTOMER_ID), 'X-Signature': signature}
# 필요한 개인정보
BASE_URL = 'https://api.naver.com'
API_KEY = '-'
SECRET_KEY = '-'
CUSTOMER_ID = '2488399'
```

```python
uri = '/keywordstool'
method = 'GET'
r = requests.get(BASE_URL + uri+'?hintKeywords={}&showDetail=1'.format('자전거'),
                 headers=get_header(method, uri, API_KEY, SECRET_KEY, CUSTOMER_ID))
```

# 함수화

```python
def keyword_master(keyword):
    # 개인정보
    BASE_URL = 'https://api.naver.com'
    API_KEY = '-'
    SECRET_KEY = '-'
    CUSTOMER_ID = '2488399'
    # 키워드 수집
    uri = '/keywordstool'
    method = 'GET'
    r = requests.get(BASE_URL + uri+'?hintKeywords={}&showDetail=1'.format(keyword),
                 headers=get_header(method, uri, API_KEY, SECRET_KEY, CUSTOMER_ID))
    df = pd.DataFrame(r.json()['keywordList'])
    return df
```

```python
keyword = "실내자전거"
df = keyword_master(keyword)
```

```python
def item_count(keyword):
    url = f"https://search.shopping.naver.com/search/all?query={keyword}"
    req = requests.get(url)
    response = TextResponse(request.url, body=request.text, encoding="utf-8")
    try:
        return response.xpath('//*[@id="__next"]/div/div[2]/div[2]/div[3]/div[1]/div[1]/ul/li[1]/a/span[1]/text()')[0].extract()
    except:
        return response.xpath('//*[@id="__next"]/div/div[2]/div/div[3]/div[1]/div[1]/ul/li[1]/a/span[1]/text()')[0].extract()
    
```

#### 3. 개별과제 : 경쟁률
- 아이템수/검색수 : 높을수록 경쟁이 심한 상품 키워드
- 수집된 데이터에서 추가로 얻을수 있는 인사이트 추가
    - 논리적인 인사이트이면 인사이트 수준에 대한 평가는 하지 않음(1개이상만)

```python
df4['competition_rate'] = df4['item_count']/df4['total_search_count']
df4
```

```python
df4_sort = df4.sort_values(by="competition_rate", ascending=False)
df4_sort
```

> 다음과 같이 봤을 떄 경쟁률비율이 높은 Keyword은 헬스 운동기구임을 알 수 있다.     
실내자전거렌탈의 경우는 아이템의 수보다 검색수가 더 많으므로 가장 수익을 내기 좋은 아이템으로 보인다.
