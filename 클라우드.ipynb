# 클라우드의 정의
- 인터넷을 통해 액세스할 수 있는 서버와 이러한 서버에서 작동하는 소프트웨어와 데이터베이스를 의미한다.
- 사용자와 기업은 클라우드 컴퓨팅을 사용하면 직접 물리적 서버를 관리하거나 자체 서버에서 소프트웨어 응용 프로그램을 실행하지 않아도 된다.
- 해외 상위 시총 주식을 가진 사업이 클라우드다
- 클라우드 서비스 시장이 굉장히 빠른 성장률을 가지고 있다.
- 앞으로도 엄청난 성장률을 보여줄 것이다.

- 코로나19 이후 비대면 사회로 급격히 전환
- MS 나델라 '2년걸릴 디지털전화 2개월 만에 이루어졌다.'
- (시험) 클라우드의 장점 : 틀을 고정해놓지 않고 빠르게 업무를 수행할 수 있다. 
- 모든 산업분야에서 전체적으로 클라우드를 사용하고 있다.

# EC2 소개
- Elastic : 탄력적인/ cloud computing (C2)
- regions -> Availability Zone -> 고가용성 HA
- 다양한 서비스가 있다.
    - 컴퓨팅 서비스
        - 가상서버서비스 EC2
            - 가상머신
            - 재구성이 가능한 컴퓨팅 리소스
            - 쉽게 확장 축소 되는 컴퓨팅용량
            - 고객 업무 영역에 따른 다양한 인스턴스 타입 제공(개발은 보통 t를 많이 사용)
            - 사용한 만큼 과금(pay-as-you-go)
            - subnet안에 vpc와 함께 있는 ebs들이 IaaS/ S3 Buckets 같은 것들이 PaaS 
        - Auto Scaling 서버 자동 확장/축소 (SDN : software defined network 자기가 알아서 트랙픽 같은것을 보고 늘렸다 줄였다한다)
        - AWS Lambda 서버리스 컴퓨팅
        
## VPC
- 고객 스스로 네트워크 정의/구성
- 하드웨어 VPN 장비를 통한 On-prem IDC와의 VPN 연결지원
- 기본적으로 인스턴스들은 사설 IP가 할당됨
- 외부 접근 가능한 동적 공인 IP 또는 고정 IP 할당가능
