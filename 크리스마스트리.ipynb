```python
fig = plt.figure()

ax = fig.add_subplot(111, projection="3d")


ax.scatter(0,0,420, c="yellow", marker="*")

k=400
Z = [i for i in range(k)]
X = [math.cos(i/5)*(k-i) for i in range(k)]
Y = [math.sin(i/5)*(k-i) for i in range(k)]
ax.scatter(X,Y,Z, c="green", marker="^")
k=400
step = 6
Z = [i for i in range(1,k,step)]
X = [math.cos(i/5+2)*(k-i+10) for i in range(1,k,step)]
Y = [math.sin(i/5+2)*(k-i+10) for i in range(1,k,step)]
ax.scatter(X,Y,Z, c="red", marker="o")

k= 100
Z = [-i*1.3+50 for i in range(k)]
X = [math.cos(i)*k for i in range(k)]
Y = [math.sin(i)*k for i in range(k)]
ax.scatter(X,Y,Z, c="chocolate", marker="^")


x = [np.random.randint(600) for _ in range(100) ]
y = [np.random.randint(600) for _ in range(100) ]
z = [np.random.randint(400) for _ in range(100) ]
ax.scatter(x,y,z, marker='o', s=15, c='grey')



plt.xlim(-700,700)
plt.ylim(-700,700)

ax.set_title("Aivler Merry Christmas !")
plt.show()
```
