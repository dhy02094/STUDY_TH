```python
# define image load function
def image_load(path):
    path = path
    os.chdir(path)
    files = os.listdir(path)
    
    data = []
    for file in files:
        f = cv2.imread(file)
        data.append(f)
        
    path_split = path.split('\\')
    
    # outlier image set
    if 'outlier' in path_split :
        return data
    
    # normal image set
    else:
        get, out = train_test_split(data,
                                   shuffle=True,
                                   test_size = 0.9, #get size = 10%
                                   random_state=42)
        return (get, out)
```
